## Collecting Data From Files

```python
import pathlib
from PIL import Image
from tqdm import tqdm
import pandas as pd

from typing import Tuple
import json
from collections import defaultdict,Counter
```

```python
train_folder = pathlib.Path('train/')
test_folder = pathlib.Path('test/')

train_images_folder = train_folder / 'images'
train_labels_folder = train_folder / 'labels'
train_targets_folder = train_folder / 'targets'

test_images_folder = test_folder / 'images'
test_labels_folder = test_folder / 'labels'
test_targets_folder = test_folder / 'targets'
```

```python
def get_image_size(path:str) -> Tuple[int,int]:
  image = Image.open(path)
  size = image.size
  image.close()
  return size
```

```python
def get_label_info(path: str) -> dict:
  info = {}
  with open(path,'r') as json_file:
    content = json.load(json_file)
    info['gsd'] = content['metadata']['gsd']
    info['capture_date'] = content['metadata']['capture_date']
    info['disaster'] = content['metadata']['disaster']
    info['disaster_type'] = content['metadata']['disaster_type']
    info['height'] = content['metadata']['height']
    info['width'] = content['metadata']['width']
    info['original_height'] = content['metadata']['original_height']
    info['original_width'] = content['metadata']['original_width']

    damage_type_count = defaultdict(lambda :0)
    feature_type_count = defaultdict(lambda :0)
    if content.get('features') is not None and content.get('features').get('xy') is not None:
      for annotation in content['features']['xy']:
        if annotation.get('properties') is not None:
          damage_type = annotation['properties'].get('subtype')
          feature_type = annotation['properties'].get('feature_type')

          if damage_type is None:
            damage_type = 'NA'
          damage_type_count[damage_type] += 1

          if feature_type is None:
            feature_type = 'NA'
          feature_type_count[feature_type] += 1
      
    info['damage_type_count'] = Counter(damage_type_count)
    info['feature_type_count'] = Counter(feature_type_count)
  
  return info
```

```python
def get_images_data(images_folder, labels_folder,targets_folder):
  data = []
  erratic = []
  for path in images_folder.glob('*.png'):
    try:
      disaster, id, time, suffix = path.stem.split('_')
      label_path = labels_folder / (path.stem + '.json')
      target_path = targets_folder / (path.stem + '_target.png')

      if label_path.exists():
        label_info = get_label_info(label_path)
      else:
        label_info = {}

      data.append({
        'disaster': disaster,
        'id': id,
        'name': path.name,
        'time':  time,
        'label_exists': label_path.exists(),
        'target_exists': target_path.exists(),
        **label_info
      })
    except Exception as e:
      erratic.append((path,e))
  return data,erratic
```

```python
train_data, erratic = get_images_data(train_images_folder,train_labels_folder,train_targets_folder)
```

```python
test_data, erratic = get_images_data(test_images_folder,test_labels_folder,test_targets_folder)
```

### Convert to pandas dataframe

for more queries we need to convert data to pandas dataframe

```python
def convert_to_dataframe(data):
  result = []
  for record in data:
    r = record.copy()
    del r['feature_type_count']

    damage_type_count = r.pop('damage_type_count')
    r['NA'] = damage_type_count.get('NA',0)
    r['destroyed'] = damage_type_count.get('destroyed',0)
    r['major-damage'] = damage_type_count.get('major-damage',0)
    r['minor-damage'] = damage_type_count.get('minor-damage',0)
    r['no-damage'] = damage_type_count.get('no-damage',0)
    r['un-classified'] = damage_type_count.get('un-classified',0)
    result.append(r)
  return pd.DataFrame(result)
```

```python
train_df = convert_to_dataframe(train_data)
test_df = convert_to_dataframe(test_data)
```

```python
train_df.to_csv('train_data_information.csv')
test_df.to_csv('test_data_information.csv')
```
